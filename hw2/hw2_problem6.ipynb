```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Bayes' theorem function
def posterior_probability(prior, sensitivity, specificity):
    false_positive_rate = 1 - specificity
    prob_positive = (sensitivity * prior) + (false_positive_rate * (1 - prior))
    return (sensitivity * prior) / prob_positive

# Define ranges
prior_range = np.linspace(0.001, 0.2, 100)  # Varying prior probability
sensitivity_range = np.linspace(0.5, 1.0, 100)  # Varying sensitivity
specificity_range = np.linspace(0.5, 1.0, 100)  # Varying specificity

# Fixed values for non-varying parameters
fixed_prior = 0.01
fixed_sensitivity = 0.95
fixed_specificity = 0.90

# Posterior probabilities
posterior_prior = [posterior_probability(p, fixed_sensitivity, fixed_specificity) for p in prior_range]
posterior_sensitivity = [posterior_probability(fixed_prior, s, fixed_specificity) for s in sensitivity_range]
posterior_specificity = [posterior_probability(fixed_prior, fixed_sensitivity, sp) for sp in specificity_range]

# Plot 1: Posterior vs Prior Probability
plt.figure(figsize=(8, 5))
plt.plot(prior_range, posterior_prior, label="Posterior Probability", color="b")
plt.xlabel("Prior Probability P(X)")
plt.ylabel("Posterior Probability P(X | Positive)")
plt.title("Effect of Prior Probability on Posterior Probability")
plt.grid(True)
plt.legend()
plt.show()

# Plot 2: Posterior vs Sensitivity
plt.figure(figsize=(8, 5))
plt.plot(sensitivity_range, posterior_sensitivity, label="Posterior Probability", color="g")
plt.xlabel("Test Sensitivity")
plt.ylabel("Posterior Probability P(X | Positive)")
plt.title("Effect of Sensitivity on Posterior Probability")
plt.grid(True)
plt.legend()
plt.show()

# Plot 3: Posterior vs Specificity
plt.figure(figsize=(8, 5))
plt.plot(specificity_range, posterior_specificity, label="Posterior Probability", color="r")
plt.xlabel("Test Specificity")
plt.ylabel("Posterior Probability P(X | Positive)")
plt.title("Effect of Specificity on Posterior Probability")
plt.grid(True)
plt.legend()
plt.show()
```
