```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
def simulate_discrete_rw(T=1000):
  """Simulates a simple discrete random walk given a time horizon T."""
  return np.cumsum(np.where(np.random.randint(0, 2, size=(T)) == 0, -1, 1), axis=0)
```

```python
# Visualize a few random walkers
num_walks = 100
time_horizon = 5000
f, ax = plt.subplots(1, 1, figsize=(10, 4))

for _ in range(num_walks):
  x = simulate_discrete_rw(T=time_horizon)
  ax.plot(x, color='maroon', alpha=0.5)
```
